### using randint()

```python
rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

#Write your code below this line ðŸ‘‡

your_choice=int(input("What do you choose? Type 0 for rock , 1 for Paper or 2 for Scissors "))
list=[rock,paper,scissors]
#printing users choice
print(your_choice)
if your_choice<3:
    print(list[your_choice])
    import random
    #printing(computers choice)
    computer_choice=random.randint(0,2)
    print("Computer chose:")
    print(computer_choice)
    print(list[computer_choice])



    if computer_choice==0 and your_choice==2:
        print("You lose")
    elif computer_choice==0 and your_choice==0:
        print("Its draw")
    elif computer_choice==0 and your_choice==1:
        print("You win")
    elif computer_choice==1 and your_choice==0:       
        print("You lose")
    elif computer_choice==1 and your_choice==1:
        print("Its draw")
    elif computer_choice==1 and your_choice==2:
        print("You win")
    elif computer_choice==2 and your_choice==0:       
        print("You win")
    elif computer_choice==2 and your_choice==1:
        print("You lose")
    elif computer_choice== 2 and your_choice==2:
        print("Its draw")    
    
else:
    print("Invalid entry")

```
